```python
from natal import Data, Stats, Chart, Config
from datetime import datetime
from IPython.display import HTML

mimi = Data(
    name = "MiMi",
    city = "Taipei",
    dt = "1980-04-20 14:30"
)

current = Data(
    name = "Current",
    city = "Taipei",
    dt = datetime.now()
)
```

```python
config = Config(theme_type="light")
chart = Chart(data1=mimi, data2=current, width=600, config=config)
HTML(chart.svg)
```

```python
stats = Stats(data1=mimi, data2=current)
HTML(stats.full_report("html"))
```
